```python
import json
import gensim
from gensim.test.utils import datapath
from gensim import models
from gensim import corpora
import numpy as np
import pandas as pd
import re
from pymystem3 import Mystem
mystem = Mystem()

from stop_words import get_stop_words

import string

import nltk
import pyLDAvis.gensim
import time
import matplotlib.pyplot as plt
from collections import Counter
```

```python
def get_texts(data, ngr:int = None):
    texts = []
    for sen in data:
        text = []
        bi_text = []
        for tok in mystem.lemmatize(sen):
            if tok not in stop_words and len(tok) > 3:
                text.append(tok)
        if ngr is not None:
            for ins in nltk.ngrams(text, ngr):
                bi_text.append(' '.join(ins))
                
        if  text:
            texts.append(text+bi_text)

    return(texts)
```

```python
file = open('/Users/aleksejfilippov/Desktop/Python_projects/data_for_topic model/a_lot_of_rus_mes.txt', 'r')
data_rus = []
for line in file.readlines():
    if len(line)>10 and len(line) < 100:
        data_rus.append(line.lower())
file.close()

file = open('/Users/aleksejfilippov/Desktop/Python_projects/data_for_topic model/stop_words.txt', 'r')
stop_words = [x[:-1] for x in file.readlines()]
file.close()
stop_words = stop_words + ['?. ','\n', '"\n',': ', '\': "\'',') ','! ',')\n', '\://', 'vk',
                           '\')))\n\'','https', 'llhhll', 'https llhhll', '', '', '', '', '', '', '', '', '', '', '', '', '', '', '', '', '', ''
                          ] + [str(i) for i in range(100)]
```

```python
count1 = Counter(" ".join(data_rus).split()).most_common(150)
df1 = pd.DataFrame.from_dict(count1)
df1 = df1.rename(columns={0: "words", 1 : "count"})
stop_words = stop_words + df1.words[:100].tolist()
# df1[:50]
```

```python
texts = get_texts(data_rus)
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
%time ldamodel = models.ldamodel.LdaModel(corpus, id2word=dictionary, num_topics=10, passes=20)
```

```python
perplexity = ldamodel.log_perplexity(list(corpus))
2**(-perplexity)
```
